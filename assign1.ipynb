```python
import pandas as pd
#Load your Uber dataset
uber_data=pd.read_csv("Uber_Drives_2016_COE.csv")
#Rename the columns
uber_data = uber_data.rename(columns={
    'START_DATE*': 'START_DATE',
    'END_DATE*': 'END_DATE',
    'START*': 'START'
})
# Display the DataFrame
uber_data
```

ASSIGNMENT-2

```python
rides=uber_data.loc[uber_data['CATEGORY*']=='Business'] # displays where category==business
rides
```

```python
longest = uber_data.sort_values(by='MILES*', ascending=False).head(5) # displays top 5 rows in miles with longest distance
longest 
```

```python
uber_data['PURPOSE*'] = uber_data['PURPOSE*'].fillna('Not Specified') # replaces all NaN values with Not Specified in PURPOSE column
uber_data['PURPOSE*'] 
```

```python
uber_data['START_DATE'] = pd.to_datetime(uber_data['START_DATE'], errors='coerce') # changing dtype to datetime
uber_data['END_DATE']   = pd.to_datetime(uber_data['END_DATE'], errors='coerce')   # changing dtype to datetime
uber_data['TRIP_DURATION'] = uber_data['END_DATE'] - uber_data['START_DATE']      # creating trip_duration and value=enddate-startdate
uber_data['TRIP_DURATION']
```

```python
sorted_trips = uber_data.sort_values(by='MILES*', ascending=False)  # sort the distance in decending order by ascending=False
sorted_trips
```
